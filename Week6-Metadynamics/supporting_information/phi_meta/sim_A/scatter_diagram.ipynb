```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取 COLVAR 文件，指定列名
col_names = ['time', 'phi', 'psi']
data = pd.read_csv("COLVAR", delim_whitespace=True, names=col_names, comment='#', skiprows=1)

# 创建二维散点图
plt.figure(figsize=(8, 6))
plt.scatter(data['phi'], data['psi'], alpha=0.5, s=10, c='blue', edgecolors='none')
plt.xlabel('Phi (radians)')
plt.ylabel('Psi (radians)')
plt.title('2D Scatter Plot of Phi vs Psi')
plt.grid(True)

# 显示图像
plt.show()
```
