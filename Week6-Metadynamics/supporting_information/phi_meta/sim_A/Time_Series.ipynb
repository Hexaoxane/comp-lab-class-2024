```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取 COLVAR 文件，指定列名
col_names = ['time', 'phi', 'psi']
data = pd.read_csv("COLVAR", delim_whitespace=True, names=col_names, comment='#', skiprows=1)

# 确认数据类型
print(data.dtypes)

# 绘制 phi 的时间序列图
plt.figure(figsize=(10, 6))
plt.plot(data['time'], data['phi'], label='phi')
plt.xlabel('Time (ps)')
plt.ylabel('Phi (radians)')
plt.title('Time Series of Phi')
plt.legend()
plt.show()

# 绘制 psi 的时间序列图
plt.figure(figsize=(10, 6))
plt.plot(data['time'], data['psi'], label='psi', color='orange')
plt.xlabel('Time (ps)')
plt.ylabel('Psi (radians)')
plt.title('Time Series of Psi')
plt.legend()
plt.show()
```
