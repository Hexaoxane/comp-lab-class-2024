```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 读取 COLVAR 文件，指定列名
col_names = ['time', 'phi', 'psi']
data = pd.read_csv("COLVAR", delim_whitespace=True, names=col_names, comment='#', skiprows=1)

# 创建二维直方图（热图）
plt.figure(figsize=(8, 6))
plt.hist2d(data['phi'], data['psi'], bins=50, cmap='viridis')
plt.colorbar(label='Count')

# 设置轴标签和标题
plt.xlabel('Phi (radians)')
plt.ylabel('Psi (radians)')
plt.title('2D Histogram of Phi and Psi Angles')

# 显示图像
plt.show()
```
