```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load fes.dat file, extracting only the first three columns
fes_data = np.loadtxt('fes.dat')[:, :3]  # Adjust path if needed

# Extract phi, psi and free energy data
phi = fes_data[:, 0]  # First column as phi values
psi = fes_data[:, 1]  # Second column as psi values
free_energy = fes_data[:, 2]  # Third column as free energy values

# Create a grid for psi and phi
grid_size = 60  # Based on earlier check of fes.dat (3600 total points)
phi_grid = np.linspace(min(phi), max(phi), grid_size)
psi_grid = np.linspace(min(psi), max(psi), grid_size)

# Reshape the free energy data to a 2D grid (60x60)
free_energy_grid = free_energy.reshape((grid_size, grid_size))

# Plot 3D free energy surface from three different angles
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
phi_grid, psi_grid = np.meshgrid(phi_grid, psi_grid)
ax.plot_surface(phi_grid, psi_grid, free_energy_grid, cmap='viridis')
ax.set_xlabel('Phi (radians)')
ax.set_ylabel('Psi (radians)')
ax.set_zlabel('Free Energy (kJ/mol)')
ax.set_title('3D Free Energy Surface (View 1)')
plt.savefig("3D_Free_Energy_Surface_View1.png")

# Second angle
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(phi_grid, psi_grid, free_energy_grid, cmap='viridis')
ax.view_init(elev=30, azim=45)  # Adjust view
ax.set_xlabel('Phi (radians)')
ax.set_ylabel('Psi (radians)')
ax.set_zlabel('Free Energy (kJ/mol)')
ax.set_title('3D Free Energy Surface (View 2)')
plt.savefig("3D_Free_Energy_Surface_View2.png")

# Third angle
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(phi_grid, psi_grid, free_energy_grid, cmap='viridis')
ax.view_init(elev=60, azim=90)  # Another view
ax.set_xlabel('Phi (radians)')
ax.set_ylabel('Psi (radians)')
ax.set_zlabel('Free Energy (kJ/mol)')
ax.set_title('3D Free Energy Surface (View 3)')
plt.savefig("3D_Free_Energy_Surface_View3.png")

# Plot 2D free energy surface with contour lines
plt.figure(figsize=(10, 8))
contour = plt.contourf(phi_grid, psi_grid, free_energy_grid, 20, cmap='viridis')
plt.colorbar(contour, label='Free Energy (kJ/mol)')
plt.xlabel('Phi (radians)')
plt.ylabel('Psi (radians)')
plt.title('2D Free Energy Surface (Phi vs Psi)')
plt.savefig("2D_Free_Energy_Surface.png")

plt.show()
```
