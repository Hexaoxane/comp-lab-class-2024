```python
import numpy as np
import matplotlib.pyplot as plt

def load_energy_data(file_path):
    """读取能量数据文件，返回能量值"""
    energy = []
    with open(file_path, 'r') as file:
        for line in file:
            # 跳过注释行
            if line.startswith('#') or line.startswith('@'):
                continue
            data = line.split()
            energy.append(float(data[1]))
    return np.array(energy)

# 加载能量数据
energy_300 = load_energy_data('T300/potential.xvg')
energy_400 = load_energy_data('T400/potential.xvg')
energy_600 = load_energy_data('T600/potential.xvg')

# 绘制能量直方图及其重叠
plt.figure(figsize=(10, 6))
plt.hist(energy_300, bins=50, alpha=0.5, label='300K', density=True)
plt.hist(energy_400, bins=50, alpha=0.5, label='400K', density=True)
plt.hist(energy_600, bins=50, alpha=0.5, label='600K', density=True)

plt.xlabel('Potential Energy (kJ/mol)')
plt.ylabel('Probability Density')
plt.title('Energy Histograms at Different Temperatures')
plt.legend()
plt.show()
```
