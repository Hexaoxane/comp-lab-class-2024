```python
import numpy as np
import matplotlib.pyplot as plt

def load_temperature_data(file_path):
    """读取温度数据文件，返回时间和温度"""
    time = []
    temperature = []
    with open(file_path, 'r') as file:
        for line in file:
            # 跳过注释行
            if line.startswith('#') or line.startswith('@'):
                continue
            data = line.split()
            time.append(float(data[0]))
            temperature.append(float(data[1]))
    return np.array(time), np.array(temperature)

# 加载不同温度下的温度数据
time_300, temp_300 = load_temperature_data('T300/temperature.xvg')
time_400, temp_400 = load_temperature_data('T400/temperature.xvg')
time_600, temp_600 = load_temperature_data('T600/temperature.xvg')

# 创建图形对象
plt.figure(figsize=(10, 6))

# 绘制不同温度的时间变化曲线
plt.plot(time_300, temp_300, label='300K', alpha=0.8)
plt.plot(time_400, temp_400, label='400K', alpha=0.8)
plt.plot(time_600, temp_600, label='600K', alpha=0.8)

# 设置图形标签和标题
plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.title('Temperature Variation Over Time at Different Temperatures')
plt.legend()

# 显示图形
plt.show()
```
