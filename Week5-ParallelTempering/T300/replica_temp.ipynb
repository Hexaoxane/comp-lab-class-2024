```python
import matplotlib.pyplot as plt
import numpy as np

# 读取数据
data = np.loadtxt('replica_temp.xvg', comments=['@', '#'])

# 提取时间和温度
time = data[:, 0]
temps = data[:, 1:]

# 绘制每个复制的温度随时间变化
plt.figure(figsize=(10, 6))
for i in range(temps.shape[1]):
    plt.plot(time, temps[:, i], label=f'Replica {i+1}')

plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.title('Temperature Variation of Replicas Over Time')
plt.legend()
plt.show()
```
