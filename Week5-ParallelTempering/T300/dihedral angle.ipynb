```python
import mdtraj as md
traj = md.load('adp.xtc', top='adp.gro')

# Extract the atom indices for the phi and psi dihedral angles
phi_indices = md.compute_phi(traj)[0]  # Ensure this is the correct indices format
psi_indices = md.compute_psi(traj)[0]  # Ensure this is the correct indices format

# Compute the dihedral angles using the correct indices
phi_angles = md.compute_dihedrals(traj, phi_indices)
psi_angles = md.compute_dihedrals(traj, psi_indices)

import matplotlib.pyplot as plt
plt.plot(phi_angles, label='Phi')
plt.plot(psi_angles, label='Psi')
plt.xlabel('Time (ps)')
plt.ylabel('Dihedral Angle (radians)')
plt.legend()
plt.show()
```
