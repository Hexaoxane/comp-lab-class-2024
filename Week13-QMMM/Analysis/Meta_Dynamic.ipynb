```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Function to read metadynamics COLVAR data
def read_metad_colvar(filepath):
    time = []
    colvar = []
    with open(filepath, 'r') as file:
        for line in file:
            if line.strip() and not line.startswith('#'):  # Skip comments
                data = line.split()
                time.append(float(data[0]))  # Column 1: Time
                colvar.append(float(data[1]))  # Column 2: Collective Variable
    return np.array(time), np.array(colvar)

# File paths
monitor_colvar_file = '/mnt/d/QM_E/Week13-QMMM/Outputs/monitor/MONITOR-COLVAR.metadynLog'
metad1_colvar_file = '/mnt/d/QM_E/Week13-QMMM/Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog'
metad2_colvar_file = '/mnt/d/QM_E/Week13-QMMM/Outputs/metad1_initmonitor_freqhills/METADYN-COLVAR.metadynLog'

# Read data
monitor_time, monitor_colvar = read_metad_colvar(monitor_colvar_file)
metad1_time, metad1_colvar = read_metad_colvar(metad1_colvar_file)
metad2_time, metad2_colvar = read_metad_colvar(metad2_colvar_file)

# Convert time to ps
monitor_time_ps = monitor_time / 1000.0
metad1_time_ps = metad1_time / 1000.0
metad2_time_ps = metad2_time / 1000.0

# Plot
plt.figure(figsize=(10, 6))
plt.plot(monitor_time_ps, monitor_colvar, label='Standard QM/MM', linestyle='-', marker='o', markersize=3)
plt.plot(metad1_time_ps, metad1_colvar, label='Metadynamics (initmonitor)', linestyle='--', marker='x', markersize=3)
plt.plot(metad2_time_ps, metad2_colvar, label='Metadynamics (freqhills)', linestyle='-.', marker='s', markersize=3)
plt.xlabel('Time (ps)')
plt.ylabel('Reaction Coordinate ()')
plt.title('Reaction Coordinate vs Time\nHill Parameters: Height=1.5 kcal/mol, Width=0.5, Freq=100 steps')
plt.legend()
plt.grid(True)

# Save plot
output_path = '/mnt/d/QM_E/Week13-QMMM/Figures/metadynamics_reaction_coordinate.png'
os.makedirs(os.path.dirname(output_path), exist_ok=True)
plt.savefig(output_path, dpi=300)
plt.show()

print(f"Plot saved to {output_path}")
```
