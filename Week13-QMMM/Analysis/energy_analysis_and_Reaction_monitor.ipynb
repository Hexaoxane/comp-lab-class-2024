```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Function to read energy files
def read_energy_file(filepath):
    time = []
    energy = []
    with open(filepath, 'r') as file:
        for line in file:
            if line.strip() and not line.startswith('#'):  # Skip comments
                data = line.split()
                time.append(float(data[1]))  # Column 2: Time (fs)
                energy.append(float(data[4]))  # Column 5: Potential Energy (a.u.)
    return np.array(time), np.array(energy)

# Define file paths
nvt_file = '/mnt/d/QM_E/Week13-QMMM/Outputs/NVT-1.ener'
npt_file = '/mnt/d/QM_E/Week13-QMMM/Outputs/NPT-1.ener'

# Check if files exist
assert os.path.exists(nvt_file), "NVT file does not exist!"
assert os.path.exists(npt_file), "NPT file does not exist!"

# Read energy data
nvt_time, nvt_energy = read_energy_file(nvt_file)
npt_time, npt_energy = read_energy_file(npt_file)

# Conversion factors
AU_TO_KJMOL = 2625.5  # Atomic units to kJ/mol
nvt_time_ps = nvt_time / 1000.0  # Convert fs to ps
nvt_energy_kjmol = nvt_energy * AU_TO_KJMOL  # Convert a.u. to kJ/mol
npt_time_ps = npt_time / 1000.0  # Convert fs to ps
npt_energy_kjmol = npt_energy * AU_TO_KJMOL  # Convert a.u. to kJ/mol

# Plot potential energy vs time
plt.figure(figsize=(10, 6))
plt.plot(nvt_time_ps, nvt_energy_kjmol, label='NVT Simulation', linestyle='-', marker='o', markersize=3)
plt.plot(npt_time_ps, npt_energy_kjmol, label='NPT Simulation', linestyle='--', marker='x', markersize=3)
plt.xlabel('Time (ps)')
plt.ylabel('Potential Energy (kJ/mol)')
plt.title('Potential Energy vs Time')
plt.legend()
plt.grid(True)

# Save the plot
output_path = '/mnt/d/QM_E/Week13-QMMM/Figures/energy_vs_time_updated.png'
os.makedirs(os.path.dirname(output_path), exist_ok=True)
plt.savefig(output_path, dpi=300)
plt.show()

print(f"Plot saved to {output_path}")
```
