```python
import os
import MDAnalysis as mda
from MDAnalysis.analysis.msd import EinsteinMSD
import matplotlib.pyplot as plt
import numpy as np

# 创建 Figures 文件夹（如果不存在）
if not os.path.exists('../Figures'):
    os.makedirs('../Figures')

# 文件路径
topology_file = "../Data/kalj_n360.data"
trajectory_files = [
    "../Data/prod/kalj_T1.5_n360_v300_prod_10000_1.unwrap.dcd",
    "../Data/prod/kalj_T1_n360_v300_prod_50000_1.unwrap.dcd",
    "../Data/prod/kalj_T0.9_n360_v300_prod_150000_1.unwrap.dcd",
    "../Data/prod/kalj_T0.8_n360_v300_prod_150000_1.unwrap.dcd",
    "../Data/prod/kalj_T0.7_n360_v300_prod_180000_1.unwrap.dcd",
    "../Data/prod/kalj_T0.65_n360_v300_prod_350000_1.unwrap.dcd",
    "../Data/prod/kalj_T0.6_n360_v300_prod_1000000_1.unwrap.dcd",
    "../Data/prod/kalj_T0.55_n360_v300_prod_3000000_1.unwrap.dcd",
    "../Data/prod/kalj_T0.5_n360_v300_prod_10000000_1.unwrap.dcd",
    "../Data/prod/kalj_T0.475_n360_v300_prod_30000000_1.unwrap.dcd"
]

# 温度列表
temperatures = [1.5, 1.0, 0.9, 0.8, 0.7, 0.65, 0.6, 0.55, 0.5, 0.475]

def get_msd_time(topology_file, trajectory_file, dt=0.005):
    try:
        # 加载 LAMMPS DATA 和 DCD 文件
        print(f"Processing file: {trajectory_file}")
        u = mda.Universe(
            topology_file,
            trajectory_file,
            format='DCD',
            topology_format='DATA',
            atom_style='id type x y z'
        )
        
        # 计算 MSD
        msd = EinsteinMSD(u, select='type 1', msd_type='xyz', fft=True)
        msd.run()

        # 获取时间和 MSD 值
        times = msd.times * dt
        msd_values = msd.results.timeseries

        # 找到 MSD = 1 的时间
        if len(msd_values) > 0:
            time_when_msd_1 = times[np.argmin(np.abs(msd_values - 1))]
            print(f"MSD=1 at time: {time_when_msd_1} ps")
            return time_when_msd_1
        else:
            print("MSD values are empty.")
            return None

    except Exception as e:
        print(f"Error processing file {trajectory_file}: {e}")
        return None

# 计算所有温度的 MSD 时间
msd_times = [get_msd_time(topology_file, traj) for traj in trajectory_files]
print("MSD times:", msd_times)

# 绘制 MSD = 1 的时间 vs 1/T 图
inverse_temp = [1.0 / T for T in temperatures]

plt.figure(figsize=(8, 6))
plt.loglog(inverse_temp, msd_times, marker='o', linestyle='--')
plt.xlabel('1/T')
plt.ylabel('Time (ps) for MSD=1')
plt.title('MSD=1 Time vs Inverse Temperature')
plt.grid(True)

# 保存图片
output_figure = '../Figures/msd_vs_temp.png'
plt.savefig(output_figure)
print(f"Figure saved as {output_figure}")
plt.show()
```

```python
import MDAnalysis as mda
from MDAnalysis.analysis.rdf import InterRDF
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import os

# 创建 Figures 目录
if not os.path.exists('../Figures'):
    os.makedirs('../Figures')

# 文件路径
topology_file = "../Data/kalj_n360.data"
trajectory_files = [
    "../Data/prod/kalj_T1.5_n360_v300_prod_10000_1.unwrap.dcd",
    "../Data/prod/kalj_T1_n360_v300_prod_50000_1.unwrap.dcd",
    "../Data/prod/kalj_T0.9_n360_v300_prod_150000_1.unwrap.dcd"
]
temperatures = [1.5, 1.0, 0.9]

# 计算 RDF
def calculate_rdf(atom_type1, atom_type2, trajectory_file):
    u = mda.Universe(topology_file, trajectory_file, format='DCD', topology_format='DATA', atom_style='id type x y z')
    rdf = InterRDF(u.select_atoms(f'type {atom_type1}'), u.select_atoms(f'type {atom_type2}'), nbins=75)
    rdf.run()
    
    # 使用 results.bins 和 results.rdf 而不是 bins 和 rdf
    return rdf.results.bins, rdf.results.rdf

# 绘制 3D RDF 图
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# 使用不同的 z 值将各个温度的数据分开
z_offset = 0
z_step = 5

for atom_type1, atom_type2, label in [(1, 1, "g_11"), (2, 2, "g_22"), (1, 2, "g_12")]:
    for traj, temp in zip(trajectory_files, temperatures):
        bins, rdf_values = calculate_rdf(atom_type1, atom_type2, traj)
        
        # 创建 3D 数据
        z = np.full_like(bins, z_offset)
        ax.plot(bins, rdf_values, z, label=f"{label}, T={temp}")
        
        # 增加 z_offset 以分隔不同温度的数据
        z_offset += z_step

ax.set_xlabel('r')
ax.set_ylabel('g(r)')
ax.set_zlabel('Temperature Offset')
ax.set_title('3D Radial Distribution Function (RDF)')
ax.legend()
plt.grid(True)

# 保存 3D 图像
plt.savefig('../Figures/rdf_3d.png')
plt.show()
```

```python
import MDAnalysis as mda
from MDAnalysis.analysis.msd import EinsteinMSD
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import os

# 创建 Figures 目录
if not os.path.exists('../Figures'):
    os.makedirs('../Figures')

# 文件路径
topology_file = "../Data/kalj_n360.data"
trajectory_files = [
    "../Data/prod/kalj_T1.5_n360_v300_prod_10000_1.unwrap.dcd",
    "../Data/prod/kalj_T1_n360_v300_prod_50000_1.unwrap.dcd",
    "../Data/prod/kalj_T0.9_n360_v300_prod_150000_1.unwrap.dcd"
]
temperatures = [1.5, 1.0, 0.9]

# 创建 3D 绘图
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# 设置 z 轴偏移
z_offset = 0
z_step = 5

# 绘制 MSD 3D 图
for traj, temp in zip(trajectory_files, temperatures):
    # 加载轨迹文件
    u = mda.Universe(topology_file, traj, format='DCD', topology_format='DATA', atom_style='id type x y z')
    
    # 计算 MSD
    msd = EinsteinMSD(u, select='type 1', msd_type='xyz', fft=True)
    msd.run()
    
    # 提取时间和 MSD 数据
    times = msd.times
    msd_values = msd.results.timeseries

    # 过滤掉零或负值
    valid_indices = (times > 0) & (msd_values > 0)
    times = times[valid_indices]
    msd_values = msd_values[valid_indices]
    
    # 检查过滤后是否有数据
    if len(times) == 0 or len(msd_values) == 0:
        print(f"No valid data for T={temp}")
        continue

    # 创建 3D 数据
    z = np.full_like(times, z_offset)
    ax.plot(np.log10(times), np.log10(msd_values), z, label=f"T={temp}")
    
    # 增加 z_offset 用于分隔不同温度的数据
    z_offset += z_step

# 设置轴标签
ax.set_xlabel('Log(Time) (ps)')
ax.set_ylabel('Log(MSD)')
ax.set_zlabel('Temperature Offset')
ax.set_title('3D MSD vs Time')
ax.legend()
plt.grid(True)

# 保存图像
plt.savefig('../Figures/msd_loglog_3d.png')
plt.show()
```

```python
import MDAnalysis as mda
from MDAnalysis.analysis.msd import EinsteinMSD
import matplotlib.pyplot as plt
import numpy as np
import os
from mpl_toolkits.mplot3d import Axes3D

# 创建 Figures 目录
if not os.path.exists('../Figures'):
    os.makedirs('../Figures')

# 文件路径
topology_file = "../Data/kalj_n360.data"
trajectory_files = [
    "../Data/prod/kalj_T1.5_n360_v300_prod_10000_1.unwrap.dcd",
    "../Data/prod/kalj_T1_n360_v300_prod_50000_1.unwrap.dcd",
    "../Data/prod/kalj_T0.9_n360_v300_prod_150000_1.unwrap.dcd"
]
temperatures = [1.5, 1.0, 0.9]

def get_msd_time(trajectory_file, temp, dt=0.005):
    try:
        u = mda.Universe(topology_file, trajectory_file, format='DCD', topology_format='DATA', atom_style='id type x y z')
        msd = EinsteinMSD(u, select='type 1', msd_type='xyz', fft=True)
        msd.run()
        times = msd.times * dt
        msd_values = msd.results.timeseries

        # 过滤掉零和负值
        valid_indices = (msd_values > 0) & (times > 0)
        times = times[valid_indices]
        msd_values = msd_values[valid_indices]

        if len(msd_values) > 0:
            # 查找 MSD = 1 的时间点
            index = np.argmin(np.abs(msd_values - 1))
            return times[index] if index < len(times) else None
    except Exception as e:
        print(f"Error processing file {trajectory_file}: {e}")
    return None

# 计算每个轨迹文件的 MSD 时间
msd_times = [get_msd_time(traj, temp) for traj, temp in zip(trajectory_files, temperatures)]
inverse_temp = [1.0 / T for T in temperatures]

# 过滤掉 None 和零值
filtered_data = [(inv_T, msd, T) for inv_T, msd, T in zip(inverse_temp, msd_times, temperatures) if msd is not None and msd > 0]
if not filtered_data:
    print("No valid data to plot.")
    exit()

inverse_temp, msd_times, temperatures = zip(*filtered_data)

# 绘制 3D 对数-对数图
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 使用对数刻度绘图
ax.scatter(np.log10(inverse_temp), np.log10(msd_times), temperatures, c=temperatures, cmap='viridis', marker='o')

# 设置标签和标题
ax.set_xlabel('log(1/T)')
ax.set_ylabel('log(Time for MSD=1)')
ax.set_zlabel('Temperature (T)')
ax.set_title('3D MSD=1 Time vs Inverse Temperature')

# 保存图像
plt.savefig('../Figures/msd_vs_temp_3d.png')
plt.show()
```
