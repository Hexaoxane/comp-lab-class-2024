```python
import matplotlib.pyplot as plt
import numpy as np

# 读取回转半径数据
data = np.loadtxt("gyration_radius.xvg", comments=["@", "#"])
time = data[:, 0]  # 时间 (ps)
gyration = data[:, 1]  # 回转半径 (nm)

# 绘制图表
plt.plot(time, gyration)
plt.xlabel('Time (ps)')
plt.ylabel('Radius of Gyration (nm)')
plt.title('Radius of Gyration vs Time')
plt.savefig('gyration_radius.png')
plt.show()
```
