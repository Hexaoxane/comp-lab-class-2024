```python
import matplotlib.pyplot as plt
import numpy as np

# 读取 Na+ 和 Cl- 的 MSD 数据
time_na, msd_na = np.loadtxt("msd_na.xvg", comments=["@", "#"], unpack=True)
time_cl, msd_cl = np.loadtxt("msd_cl.xvg", comments=["@", "#"], unpack=True)

# 绘制 MSD 曲线
plt.plot(time_na, msd_na, label='Na+ MSD')
plt.plot(time_cl, msd_cl, label='Cl- MSD')
plt.xlabel('Time (ps)')
plt.ylabel('MSD (nm^2)')
plt.title('Mean Squared Displacement of Na+ and Cl-')
plt.legend()
plt.savefig('msd_na_cl_plot.png')
plt.show()
```
