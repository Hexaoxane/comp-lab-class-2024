```python
import matplotlib.pyplot as plt
import numpy as np

# 读取端到端距离数据
data = np.loadtxt("end_to_end_dist.xvg", comments=["@", "#"])
time = data[:, 0]  # 时间 (ps)
distance = data[:, 1]  # 端到端距离 (nm)

# 绘制图表
plt.plot(time, distance)
plt.xlabel('Time (ps)')
plt.ylabel('End-to-End Distance (nm)')
plt.title('End-to-End Distance vs Time')
plt.savefig('end_to_end_distance.png')
plt.show()
```
